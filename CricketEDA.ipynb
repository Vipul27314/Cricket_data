```python
plt.figure(figsize=(6, 4))
sns.countplot(y='location', hue='match_type', data=df)
plt.title('Match Types across location')
plt.xlabel('match_type')
plt.ylabel('location')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='toss_won', hue='toss_decision', data=df)
plt.title('Toss_won across Toss_decision')
plt.xlabel('toss_won')
plt.ylabel('Frequency') 
plt.legend(title='Toss_Decision', loc='upper left')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
sns.countplot(x='winner', hue='venue', data=df)
plt.title('venue across winner')
plt.xlabel('winner')
plt.ylabel('Frequency') 
plt.legend(title='venue', loc='upper left')
plt.show()
```

```python
plt.figure(figsize=[10,4])
plt.hist(df["venue"],bins=30)
plt.xlabel("venue")
plt.ylabel("Frequency")
plt.title("venue across match")
plt.xticks(rotation=90)
plt.show()
```

```python
runs = df.groupby(['team1'], as_index=False)['winner_runs'].sum().sort_values(by='winner_runs', ascending=False)
sns.set(rc={'figure.figsize':(12,6)})
plt.title("winner_runs across team1")
plt.xticks(rotation=50)
sns.barplot(data = runs, x = 'team1',y= 'winner_runs')
```

```python
runs = df.groupby(['team2'], as_index=False)['winner_runs'].sum().sort_values(by='winner_runs', ascending=False)
sns.set(rc={'figure.figsize':(12,6)})
plt.title("winner_runs across team2")
plt.xticks(rotation=50)
sns.barplot(data = runs, x = 'team2',y= 'winner_runs')
```

```python
runs = df.groupby(['team1'], as_index=False)['winner_wickets'].sum().sort_values(by='winner_wickets', ascending=False)
sns.set(rc={'figure.figsize':(12,6)})
plt.title("winner_wickets across team1")
plt.xticks(rotation=50)
sns.barplot(data = runs, x = 'team1',y= 'winner_wickets')
```

```python
runs = df.groupby(['team2'], as_index=False)['winner_wickets'].sum().sort_values(by='winner_wickets', ascending=False)
sns.set(rc={'figure.figsize':(12,6)})
plt.title("winner_wickets across team2")
plt.xticks(rotation=50)
sns.barplot(data = runs, x = 'team2',y= 'winner_wickets')
```
